```python
import os
import pickle


actors = os.listdir('/content/drive/MyDrive/celebrity/data')
print(actors)

filenames =[]

for actor in actors:
    for file in os.listdir(os.path.join('/content/drive/MyDrive/celebrity/data',actor)):
        filenames.append(os.path.join('/content/drive/MyDrive/celebrity/data',actor,file))

pickle.dump(filenames,open('/content/drive/MyDrive/celebrity/filenames.pkl','wb'))

```

```python
from fileinput import filename
from xml.etree.ElementInclude import include
from tensorflow.keras.preprocessing import image
from keras_vggface.vggface import VGGFace
from keras_vggface.utils import preprocess_input
import numpy as np
import pickle

filenames = pickle.load(open('/content/drive/MyDrive/celebrity/filenames.pkl','rb'))

model = VGGFace(model='resnet50',include_top=False,input_shape=(224,224,3),pooling='avg')

print(model.summary())
```

```python
def feature_extractor(img_path,model):
  img =image.load_img(img_path,target_size=(224,224))
  img_array = image.img_to_array(img)
  expanded_img = np.expand_dims(img_array,axis=0)
  preprocessed_img = preprocess_input(expanded_img)

  result = model.predict(preprocessed_img).flatten()

  return result

```

```python
pip install tqdm
```

```python
from tqdm import tqdm
features=[]
for file in tqdm(filenames):
  features.append(feature_extractor(file,model))

pickle.dump(features,open('/content/drive/MyDrive/celebrity/embedding.pkl','wb'))
```
